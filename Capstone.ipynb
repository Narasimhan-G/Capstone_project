```
# Converting mixed data types to string before applying Label Encoder
x_train['Smoker?'] = x_train['Smoker?'].astype(str)
x_test['Smoker?'] = x_test['Smoker?'].astype(str)

x_train['Living in?'] = x_train['Living in?'].astype(str)
x_test['Living in?'] = x_test['Living in?'].astype(str)

x_train['Any heriditary condition?'] = x_train['Any heriditary condition?'].astype(str)
x_test['Any heriditary condition?'] = x_test['Any heriditary condition?'].astype(str)

# Applying the previously written Label Encoder
from sklearn.preprocessing import LabelEncoder
labelEncoder_smoker = LabelEncoder()
labelEncoder_living = LabelEncoder()
labelEncoder_heriditary = LabelEncoder()

# Fit the labelEncoders on the combined train and test data
combined_smoker = pd.concat([x_train['Smoker?'], x_test['Smoker?']])
labelEncoder_smoker.fit(combined_smoker)

combined_living = pd.concat([x_train['Living in?'], x_test['Living in?']])
labelEncoder_living.fit(combined_living)

combined_heriditary = pd.concat([x_train['Any heriditary condition?'], x_test['Any heriditary condition?']])
labelEncoder_heriditary.fit(combined_heriditary)

# Transform the train and test data
x_train['Smoker?'] = labelEncoder_smoker.transform(x_train['Smoker?'])
x_test['Smoker?'] = labelEncoder_smoker.transform(x_test['Smoker?'])

x_train['Living in?'] = labelEncoder_living.transform(x_train['Living in?'])
x_test['Living in?'] = labelEncoder_living.transform(x_test['Living in?'])

x_train['Any heriditary condition?'] = labelEncoder_heriditary.transform(x_train['Any heriditary condition?'])
x_test['Any heriditary condition?'] = labelEncoder_heriditary.transform(x_test['Any heriditary condition?'])
```

```
#Scaling the values using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
```

```
y_test = pd.read_csv('/work/output_testData.csv')
```

```
#Building the Classifier model
#Using KNN classifier
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(x_train_scaled,y_train)
y_pred = neigh.predict(x_test_scaled)
```

```
#computing accuracy_score for KNN
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```
#Using Decision tree classifier
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth=25,max_features='log2')
clf.fit(x_train_scaled,y_train)
dtclf = clf.predict(x_test_scaled)
accuracy_score(y_test,dtclf)
```

```
#Using Naive_Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB(var_smoothing=1e-5)
nb.fit(x_train_scaled,y_train)
nbclf = nb.predict(x_test_scaled)
accuracy_score(y_test,nbclf)
```

```
#Using svm classifer
from sklearn.svm import SVC
svc = SVC(kernel= 'sigmoid')
svc.fit(x_train_scaled,y_train)
svclf = svc.predict(x_test_scaled)
accuracy_score(y_test,svclf)
```

```
#Using ensemble methods
from sklearn.ensemble import AdaBoostClassifier
adb = AdaBoostClassifier(n_estimators=43, random_state=0,learning_rate=1.4)
adb.fit(x_train_scaled,y_train)
adbclf = adb.predict(x_test_scaled)
accuracy_score(y_test,adbclf)
```
